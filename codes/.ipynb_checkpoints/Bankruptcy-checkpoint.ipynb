# Data Preparation

## Removing Duplicate Rows

```python
train_v1 = train.drop_duplicates()
train_v1.shape
```

## Outlier Treatment

```python
# Capping and Flooring between three stdevs of variable mean
def emp_cap_floor(df,var):
    mean = df[var].mean()
    std = df[var].std()
    df.loc[df[var] > (mean +3*std),var] = mean + 3*std
    df.loc[df[var] < (mean -3*std),var] = mean - 3*std
    return df[var]
```

```python
# Loop for all the variables
var_list = train_v1.columns.drop('class')
for i in var_list:
    train_v1[i] = emp_cap_floor(train_v1,i)
```

## Missing Value Imputations

```python
miss_df = pd.DataFrame(train.isna().sum()).reset_index().rename(columns={'index':'Variable',0:'Num_Miss'})
print(miss_df[miss_df['Num_Miss'] > 0 ].shape[0],"missing values")
```

```python
train_v1.head()
```

## Variable Transformations

```python
# Kurtosis Calculation
from scipy.stats import kurtosis,skew
kur_list = []
skew_list = []
norm_list = []
min_val_list = []
for var in var_list:
    kur_list.append(kurtosis(train_v1[var]))
    skew_list.append(skew(train_v1[var]))
    min_val_list.append(train_v1[var].min())
ksn_df = pd.DataFrame(data = zip(var_list,kur_list,skew_list,min_val_list),columns=['Var','Kurtosis','Skewness',"Min"])
ksn_df.head()
```

```python
# Select a cutoff for skewness
sns.distplot(ksn_df['Skewness'],color='r',kde=True).set_xlim([min(ksn_df['Skewness']),max(ksn_df['Skewness'])])
std_trans_list = list(ksn_df[((ksn_df['Skewness'] > 15) | (ksn_df['Skewness'] < -15))]['Var'])
```

```python
std_trans_list
```

```python
ksn_df[ksn_df.Var.isin(std_trans_list)]
```

```python
# Before Transformation
fig, axes = plt.subplots(4,4,figsize=(38,28))
for r in range(0,4):
    for c in range(0,4):
        var = std_trans_list[r*4 + c]
        sns.distplot(train_v1[var],color='r',kde=True,ax=axes[r,c]).set_xlim([min(train_v1[var]),max(train_v1[var])])
```

- Looking at the skewness values and histograms, the variables **Attr20', 'Attr21', 'Attr23', 'Attr27', 'Attr30', 'Attr32', 'Attr37', 'Attr41', 'Attr43', 'Attr44', 'Attr47', 'Attr49', 'Attr52', 'Attr56', 'Attr58', 'Attr60', 'Attr62'** need  transformations

```python
train_v1[std_trans_list].head()
```

```python
# Taking the log transformation
import math
for tvar in std_trans_list:
    train_v1[tvar] = train_v1[tvar].apply(lambda x:math.log(5+x))
```

# Class Imbalance Problem

```python
# Class variable frequencies
sns.countplot(x='class',data=train)
plt.show()
train_v1['class'].value_counts(normalize=True)
```

- Only 2.2% of the firms are bankrupted so we have a class imbalance problem and let us use sample minority over sampling technique (SMOTE) and balance the classes in the data

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
```

```python
# Splitting the X & y
x = train_v1.loc[:,train_v1.columns!='class']
y = train_v1.loc[:,train_v1.columns=='class']

# Random Split
x_train, x_test, y_train, y_test = train_test_split(x, y.values.ravel(), test_size=0.3, random_state=123)

# Applying SMOTE
smt_obj = SMOTE(random_state=123)
os_x,os_y = smt_obj.fit_resample(x_train,y_train)
np.bincount(os_y)
```

# Feature Selection

```python
from sklearn.feature_selection import RFE
logreg = LogisticRegression()
rfe = RFE(logreg)
rfe = rfe.fit(os_x,os_y)
print(rfe.support_)
print(rfe.ranking_)
```

```python
import xgboost as xgb
```

```python
x.columns
```

```python
x = os_x.iloc[:,rfe.support_]
y = os_y
xg_logit = xgb.XGBClassifier(base_score=0.02, 
                             booster='gbtree',
                             gamma=0.5,
                             learning_rate=0.1,
                             max_depth=5,
                             min_child_weight=1,
                             n_estimators=400,
                             objective='binary:logistic',
                             random_state=123,
                             reg_lambda=1,
                             scale_pos_weight=1,
                             subsample=1,
                             verbosity=0,
                             eval_metric='auc')

xg_logit = xg_logit.fit(x,y)
```

```python
xtest = x_test.iloc[:,rfe.support_]
```

```python
y_pred = xg_logit.predict(xtest)
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test,y_pred)
print(confusion_matrix)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import roc_auc_score,roc_curve
roc = roc_auc_score(y_test,y_pred)
print("AUC = ",roccurve)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
fpr,tpr,thresholds = roc_curve(y_test,xg_logit.predict_proba(xtest)[:,1])
plt.figure()
plt.plot(fpr,tpr,[0,1],[0,1],'r--')
plt.xlim([-0.01,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristics')
plt.show()
```
